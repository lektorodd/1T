Ser me på dei første tre trekantala, kan me sjå at dei utgjer halvparten av dei første *rektangeltala* (med sider $n$ og $n+1$)

<div>

Formelen for trekanttala blir då
$$ T_n = \frac{n(n+1)}{2} $$

```python
# Hentar inn figurnummer
n = int(input("Skriv inn figurnummer: "))

# Reknar ut kuler
kuler = (n*(n+1))/2

#Skriv ut resultatet
print("Det er", int(kuler), "kuler i trekanttal nummer", n)
```

Her har eg brukt `int(kuler)` sidan antall kuler i trekanttala er heiltal.
