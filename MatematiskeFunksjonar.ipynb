# Matematiske funksjonar

Funksjonar i Python er ikkje det samme som matematiske funksjonar. I Python er dei fleste kommandoane me brukar funksjonar. Døme på dette kan vera `print()` og `float()`. 

Her skal me sjå på korleis me kan definera matematiske funksjonar i Python. 

## Polynomfunksjon

Me skal sjå på funksjonen
$$f(x)=\frac{1}{2}x^3 - \frac{3}{2}x^2 + 2, D_f = [-2,4]$$
Først skal me laga funksjonen i Python før me skal plotta han. 

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return 0.5*x**3 - 1.5*x**2 + 2

x = np.linspace(-2,4,100)
y = f(x)

# Pynte plottet
plt.xlabel("x", fontsize = 14)
plt.ylabel("y", fontsize = 14)
plt.axhline(y=0, color="black")
plt.axvline(x=0, color="black")
plt.grid()

plt.plot(x,y)

plt.title("Plot av f(x)", fontsize = 14)
plt.show()
```

```python
# Eksponentialfunksjon

def g(x):
    return 400*1.05**x

x2 = np.linspace(0,31,100)
y2 = g(x2)

# Pynte plottet
plt.xlabel("x", fontsize = 14)
plt.ylabel("y", fontsize = 14)
plt.axhline(y=0, color="black")
plt.axvline(x=0, color="black")
plt.grid()

plt.plot(x2,y2)

plt.title("Plot av g(x)", fontsize = 14)
plt.show()
```
