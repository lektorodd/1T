# Matematiske funksjonar

Funksjonar i Python er ikkje det samme som matematiske funksjonar. I Python er dei fleste kommandoane me brukar funksjonar. 
Døme på dette kan vera `print()` og `float()`. 

Her skal me sjå på korleis me kan definera matematiske funksjonar i Python. 

## Polynomfunksjon

Me skal sjå på funksjonen
$$f(x)=\frac{1}{2}x^3 - \frac{3}{2}x^2 + 2, \qquad D_f = [-2,4]$$
Først skal me laga funksjonen i Python før me skal plotta han. 

Mykje av koden me bruker her kan de kopiera og lima inn i seinare program, så **pass på at de har kontroll på kor de lagrar fila**. Gjerne i mattemappa som de laga i starten av året.

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return 0.5*x**3 - 3/2 *x**2 + 2

x1 = np.linspace(-2, 4, 100)
y1 = f(x1)

# pynte plottet
plt.xlabel("x", fontsize = 14) #namn på akse
plt.ylabel("y", fontsize = 14) #namn på akse

plt.axhline(y=0, color = "black") # x-akse
plt.axvline(x=0, color = "black") # y-akse

plt.grid() #rutenett

plt.plot(x1, y1)  #sjølve plottinga av grafen

plt.title("Plot av funksjonen f(x)", fontsize = 16)  #tittel

plt.show()  #vise det ferdige plottet

```

## Eksponentialfunksjon

Her ser me på funksjonen
$$g(x) = 2,5 \cdot 1,5^x, \qquad D_f = [-4,6]$$

*Hugs at me kan kopiera mykje av koden frå det me gjorde over*

```python
# Ein eksponentialfunksjon her

import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return 2.5 * 1.5**x

x2 = np.linspace(-4, 6, 100)
y2 = f(x2)

# pynte plottet
plt.xlabel("x", fontsize = 14) #namn på akse
plt.ylabel("y", fontsize = 14) #namn på akse

plt.axhline(y=0, color = "black") # x-akse
plt.axvline(x=0, color = "black") # y-akse

plt.grid() #rutenett

plt.plot(x2, y2)  #sjølve plottinga av grafen

plt.title("Plot av funksjonen f(x)", fontsize = 16)  #tittel

plt.show()  #vise det ferdige plottet
```
