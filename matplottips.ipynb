# Tips matplotlib.pyplot

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return 0.5*x**2 - 2

x = np.linspace(-5, 5, 100)
y = f(x)

plt.figure(figsize = (12,7))                            # <--- for å endra storleik på figuren (bredde, høgde)
plt.plot(x,y)
plt.savefig("eit-eller-anna-filnamn.png", dpi = 72)     # <--- for å lagra figuren som bilete
plt.title("Mitt plot \n", fontsize = 24)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
